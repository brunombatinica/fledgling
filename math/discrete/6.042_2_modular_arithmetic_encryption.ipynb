```python
import numpy as np
import matplotlib as mpl
from tqdm import tqdm
from time import sleep

import math

def is_prime(n):
  for i in range(2,int(math.sqrt(n))+1):
    if (n%i) == 0:
      return False
  return True


# from 6.042euclidsalgo
def gcd(x,y):
    #retruns the greatest common divisor of x and y

    #recursive implementation

    #sort numbers for more robustness
    a = min(x,y)
    b = max(x,y)

    if a == 0:
        return b
    else: 
        return gcd(b%a,a)
```

```python
def encode(input_string):
    #converts word to numbers and adds 2 digits to make it a prime - jsut by brute force
    # input : string of letters
    # output : prime number representation - each 2 digits corresponds to a letter and last 2 digits are 2 make a primer

    
    output = ""
    for i in input_string:
        temp = str(ord(i)-96)
        if len(temp) == 1:
            temp = "0" + temp
        output += temp

    # adding numbers to make prime burte force
    # start at 10 to not dealt with single digit problems
    for i in tqdm(range(11,99)):
        
        test = output + str(i)

        if is_prime(int(test)):
            return int(test)

        sleep(0.1)
        
        
    return "NO PRIME ADDITION"


def decode(input_number):
    #converts the output number  back to a string 
    # removes last to digits


    s = str(input_number)[:-2]
    
    output = []

    for i in range(len(s)//2):
        output.append(chr(int(s[(i*2):((i*2)+2)]) +96))

    return output

```

# Congruence & Relative prime


## **Prelative prime**

prelative prime iff GCD(a,b) = 1

We showed that the GCD must be a liear compination of a and b so 1 must be a linear combination of a and b if they are relatively prime

*there exists s and t such that $sa + tb = 1$*

so we can say that $sa = 1 \pm tb $

so we can kind of say that sa = 1 (hmmmmmm), and thus can view a as a multiplicative inverse of s (okay now we are going somewhere)

## **CONGRUENCE**

x is congruent to y modulo n : $ x \equiv y (mod n) $

$ iff ~ n|(x-y) $

n divides the difference between x and y 

e.g. $31 \equiv 16~(mod~5) $

as 31-16 = 15 which can be divided by n 

THINK OF A CLOCK - mod 12 - 3 and 15 are congruent



## *now we can extend congruence to get a new idea of a multiplicative inverse

modular extension of multiplicative inverse

exists iff a and n are coprime

multiplicative inverse of x mod n is a number $x^{-1}$ in $\{0,1,...,n-1\}$ s.t. 

$xx^{-1} \equiv 1 ~ (mod~n) $

e.g.

$2*3 \equiv 1 (mod ~5) $

$2 \equiv 3^{-1}~(mod~5) $

these are symetrical - so ....

$3 \equiv 2^{-1}~(mod~5) $ makes sense


a number can be its own multiplicative inverse
