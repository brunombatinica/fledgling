```python
def mon_len(mon,year):
    if mon in [1,3,5,7,8,10,12]:
        return 31
    elif mon in [4,6,9,11]:
        return 30
    else:
        if year % 4 == 0:
            if year % 100 == 0:
                if year % 400 == 0:
                    return 29
                else:
                    return 28
            else:
                return 29
        else:
            return 28
```
