```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D (Sigmoid)

```python
# normal distribution

def gauss(x,mu = 0,sd = 1):
    return (1 / (sd * np.sqrt(2 * np.pi))) * np.exp( -0.5 * ((x - mu) / sd)**2)
```

```python
#plot normal distributions
x = np.arange(-5,5,.1)
y1 = [gauss(i,0,1) for i in x]
y2 = [gauss(i,1,1) for i in x]

#compare probability of being in either 
P1 = [b/(a+b) for a,b in zip(y1,y2)]
```

```python
#create 2 normal distributions
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,P1)
```
