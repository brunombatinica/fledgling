# Gradient and Directional derivative

code from directional_derivative .py scripts

Vector of partial derivative

```python
# creating a simple 3D surface
x = np.arange(-2,2,0.1)
y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(x,y)

def f(x,y):
    z = x**2 + y**2
    return z

Z = f(X,Y) #element wise operations

#work out gradient for a specific point
g_spot = (0.5,0.5)

#gradient calculator

def grad_f(x,y):
    dfdx = 2*x #df/fx
    dfdy = 2*x #df/dy

    return np.array([dfdx,dfdy])

grad = grad_f(*g_spot)

#plane calculator


#plotting
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d") # need to set projection to be 3D
surf = ax.plot_surface(X,Y,Z,color = "blue", linewidth=0, alpha = 0.5)
grad = ax.quiver(*grad,0,*g_spot,0,color = "green")
```
