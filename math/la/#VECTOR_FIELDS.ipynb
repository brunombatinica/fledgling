```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
x = np.arange(-5,5,1)
y = np.arange(-5,5,1)
X,Y = np.meshgrid(x,y)

Fi = -Y
Fj = X

fig, ax = plt.subplots()
ax.quiver(X,Y,Fi,Fj)
```

```python
deltat = 0.05
t = np.arange(0,1,deltat)

x = t
y = t**2

dx = 1
dy = 2*t

fi = -y
fj = x

w = (dx * fi + dy * fj) * deltat

fig, ax = plt.subplots()
ax.plot(x,y,"b.")
#path derivative
ax.quiver(x,y,dx,dy,color = "red")
#field
ax.quiver(x,y,fi,fj, color = "blue")

sum(w)
```

```python
deltat = 0.05
t = np.arange(0,np.pi/4,deltat)

x = np.cos(t)
y = np.sin(t)

dx = -np.sin(t)
dy = np.cos(t)

fi = y
fj = x

w = (dx * fi + dy * fj) * deltat

fig, ax = plt.subplots()
ax.set_xlim(0,1.5)
ax.set_ylim(0,1.5)
plt.gca().set_aspect('equal', adjustable='box')
ax.plot(x,y,"b.")
#path derivative
ax.quiver(x,y,dx,dy,color = "red")
#field
ax.quiver(x,y,fi,fj, color = "blue")

sum(w)
```

```python
deltat = 0.001
t = np.arange(0,(0.5**0.5),deltat)

x = t
y = t

dx = 1
dy = 1

fi = y
fj = x

w = (dx * fi + dy * fj) * deltat

fig, ax = plt.subplots()
ax.set_xlim(0,1.5)
ax.set_ylim(0,1.5)
plt.gca().set_aspect('equal', adjustable='box')
ax.plot(x,y,"b.")
#path derivative
ax.quiver(x,y,dx,dy,color = "red")
#field
ax.quiver(x,y,fi,fj, color = "blue")

sum(w)
```

```python
#Gradient field test

# saddle function

x = np.arange(-5,5,1)
y = np.arange(-5,5,1)
X,Y = np.meshgrid(x,y)

#function stuff
Z = X**2 - Y**2

#M = fx
M = 2*X
#N = fy
N = -2*Y

#2nd derivatives
My = 0
Nx = 0

fig = plt.figure()
ax3d = fig.add_subplot(121,projection = "3d")
ax3d.plot_surface(X,Y,Z)

#gradient field
ax1 = fig.add_subplot(122)
ax1.quiver(X,Y,M,N)

```

```python
#Gradient field test

# more complicated function 

x = np.arange(-5,5,1)
y = np.arange(-5,5,1)
X,Y = np.meshgrid(x,y)

#function stuff
Z = X*Y + 5*np.sin(X*Y**2)

#M = fx
M = 2*X
#N = fy
N = -2*Y

#2nd derivatives
My = 0
Nx = 0

fig = plt.figure()
ax3d = fig.add_subplot(121,projection = "3d")
ax3d.plot_surface(X,Y,Z)

#gradient field
ax1 = fig.add_subplot(122)
ax1.quiver(X,Y,M,N)
```
