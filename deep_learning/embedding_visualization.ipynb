```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml

```

```python
# Load data
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data / 255.0
y = mnist.target
```

```python
# Convert labels to integers
y = y.astype(np.int64)

# Select a subset to speed up the process
n_samples = 3000
X_subset = X[:n_samples]
y_subset = y[:n_samples]
```

# t-SNE

```python
# t-SNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
X_subset_tsne = tsne.fit_transform(X_subset)
```

```python
# Visualization
plt.figure(figsize=(10, 10))
for i in range(10):
    indices = y_subset == i
    plt.scatter(X_subset_tsne[indices, 0], X_subset_tsne[indices, 1], label=i)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cm import get_cmap
# Visualization
plt.figure(figsize=(12, 10))
# Get the viridis colormap
cmap = get_cmap('viridis')

# Sample three colors from the colormap at maximally spaced intervals
colors = [cmap(i) for i in np.linspace(0, 1, 3)]

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Plot the data with the sampled colors
plt.scatter(x, y1, color=colors[0], label='Diagnoses and operations')
plt.scatter(x, y2, color=colors[1], label='Pharmaceuticals')
plt.scatter(x, y3, color=colors[2], label='Laboratory tests')

# Add the legend
plt.legend(loc='best')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Embedding visualization')

# Show the plot
plt.show()
```
