# The rational behind statistical modelling

Let X1,....,Xn be n independent copies of X

The goal of statistics is to **learn the distribution of X**!




If x is {0,1} then this is easy - this must be a bernoulli random variable and the best estimate of p is the mean(X)

But what if there are more options - then you need to collect more "parameters"

e.g. number of siblings



Could be one parameter for every possible option 

Described using a probablity mass function (mass for discrete)

Or you could make some assumptions which will DECREASE THE NUMBER OF PARAMETERS you have to learn

E.g. if you assume X is Poiss(lambda) Then you only have to learn one parameter

Statisticall modelling is trying to "compress" information

i.i.d is a very important assumption

If you dont assume things are much more complicated

So start with

$$ X1,....Xn ~are ~random ~variables ~i.i.d ~with ~dist~ P $$

each assumption reduces the degrees of freedom in hte space you are searching in
