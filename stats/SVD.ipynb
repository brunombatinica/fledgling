SVD is a matrix decomposition method - for reducing a matrix to its constituent pares to make certain subseqeunce calculation simpler

Ignore complex numbers currently


$ A = U \Sigma V^{T} $

$A$ = m x n matrix we want to decompose

_

$U$ = m x m matrix (Complex unitary matrix) special case where A is real then U is orthoganal (orthoganal is unitary)!

$\Sigma$ = m x n diagonal matrix

$V^{T}$ = transpose of an n x n matrix (also orthogonal)


_

Diagonal values of the $\Sigma$ matrix are known as the **SINGULAR VALUES** of the orginal matrix A

Columns of $U$ are called the **LEFT SINGULAR VECTORS** of A

Columns of $V$ are called the **RIGHT SINGULAR VECTORS** OF A
